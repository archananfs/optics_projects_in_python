```python
import scipy.special
import numpy as np
import matplotlib.pyplot as plt
```

```python
w0 = 2 # Enter the beam size in mm 
p = 0  # p and q represent specific mode of the laser
q = 1

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X,Y = np.meshgrid(x, y)
u_y = np.sqrt(2)*Y/w0
u_x = np.sqrt(2)*X/w0
p_x = scipy.special.eval_hermite(p,u_x)
p_y = scipy.special.eval_hermite(q,u_y)
H_xy =abs(p_x)*abs(p_y)*np.exp(-(X**2+Y**2)/(w0)**2)

ct = plt.contourf(H_xy,200, cmap='hot')

plt.colorbar(ct)
plt.axis('off')

plt.show()
```
